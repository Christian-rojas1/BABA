### **------------ Razones Financieras -----------**

```python
simbolo = 'BABA'
ticker_object = yf.Ticker(simbolo)
esf =ticker_object.balance_sheet
eri = ticker_object.incomestmt
cash=ticker_object.cash_flow

Activo='Total Assets'
Pasivo='Total Liabilities Net Minority Interest'
Patrimonio='Total Equity Gross Minority Interest'
Deuda='Total Debt'
Ingresos='Total Revenue'
Un='Net Income From Continuing Operation Net Minority Interest'
Inventarios='Inventory'
Gastosfin='Interest Expense'
Costoventas='Cost Of Revenue'
Cxc='Receivables'
Proveedores='Payables'
Compras=''#no esta
Activocorr='Current Assets'
Pasivocorr='Total Non Current Liabilities Net Minority Interest'
Utilidadesretenidas='Retained Earnings'
Impuestos='Tax Provision'
Accionesc='Share Issued'
Uai='Pretax Income'

Activov= pd.DataFrame(esf.loc[Activo])
Pasivov=pd.DataFrame(esf.loc[Pasivo])
Patrimoniov=pd.DataFrame(esf.loc[Patrimonio])
Deudav=pd.DataFrame(esf.loc[Deuda])
Ingresosv=pd.DataFrame(eri.loc[Ingresos])
Unv=pd.DataFrame(eri.loc[Un])
Inventariosv=pd.DataFrame(esf.loc[Inventarios])
Gastosfinv=pd.DataFrame(eri.loc[Gastosfin])
Costoventasv=pd.DataFrame(eri.loc[Costoventas])
Cxcv=pd.DataFrame(esf.loc[Cxc])
Proveedoresv=pd.DataFrame(esf.loc[Proveedores])
Activocorrv=pd.DataFrame(esf.loc[Activocorr])
Pasivocorrv=pd.DataFrame(esf.loc[Pasivocorr])
Utilidadesretenidasv=pd.DataFrame(esf.loc[Utilidadesretenidas])
Impuestosv=pd.DataFrame(eri.loc[Impuestos])
Accionescv=pd.DataFrame(esf.loc[Accionesc])
Uaiv=pd.DataFrame(eri.loc[Uai])


ratios = pd.concat([Activov, Pasivov,Patrimoniov,Deudav,Ingresosv,Unv,Inventariosv,Gastosfinv,Costoventasv,Cxcv,Proveedoresv,Activocorrv,Pasivocorrv], axis=1)
ratios.columns = ['Activo', 'Pasivo','Patrimonio','Deuda','Ingresos','Utilidad Neta','Inventarios','Gastos Financieros','Costo de Ventas','Cuentas por cobrar','Proveedores','Activo Corriente','Pasivo Corriente']
ratios['R Endeudamiento'] = ratios['Deuda'] / ratios['Patrimonio']
ratios['R Solvencia'] = ratios['Activo'] / ratios['Pasivo']
ratios['R Totacion de Activos'] = ratios['Ingresos'] / ratios['Activo']
ratios['R Margen de Utilidad'] = ratios['Utilidad Neta'] / ratios['Ingresos']
ratios['R ROA'] = ratios['Utilidad Neta'] / ratios['Activo']
ratios['R ROE'] = ratios['Utilidad Neta'] / ratios['Activo']
ratios['R Rot Inventarios'] = ratios['Costo de Ventas'] / ratios['Inventarios']
ratios['R Periodo Inventarios'] = (ratios['Inventarios'] / ratios['Costo de Ventas'])*365
ratios['R Rotacion Cobranza'] = ratios['Ingresos'] / ratios['Cuentas por cobrar']
ratios['R Liquidez'] = ratios['Activo Corriente'] / ratios['Pasivo Corriente']
ratios['R Acidez'] = (ratios['Activo Corriente'] - ratios['Inventarios'])/ratios['Pasivo Corriente']



ratios=ratios.transpose()
```

### **------------ Altman Index -----------**

```python
fechas = list(esf.columns.values)
fechas = pd.to_datetime(fechas)
fechas = fechas.tz_localize('UTC')  # Asignar la zona horaria UTC a las fechas
fechas= fechas.tz_convert(None)
fechas = fechas.date

ffin = max(fechas)
ffin = ffin + pd.Timedelta(8, 'd')
finicio = min(fechas)

pcaccio= ticker_object.history(start=finicio,end=ffin,period="1y")
pcaccio.index = pcaccio.index.tz_convert(None).date

pcaccion = pcaccio["Close"]

fecha1 = fechas[0]
fecha2 = fechas[1]
fecha3 = fechas[2]
pcaccion_fecha1 = pcaccion[pcaccion.index == fecha1]
pcaccion_fecha2 = pcaccion[pcaccion.index == fecha2]
pcaccion_fecha3 = pcaccion[pcaccion.index == fecha3]

fprecio = {
    fechas[0]: pcaccion_fecha1.values[0], #pd.Timedelta(3, 'd') #PARA ACERCAR A EL DATO MAS CERCANO
    fechas[1]: pcaccion_fecha2.values[0],
    fechas[2]: pcaccion_fecha3.values[0],
}

fprecio = pd.DataFrame.from_dict(fprecio, orient='index', columns=['Precio'])

altman=pd.concat([Activocorrv,Activov,Pasivocorrv,Pasivov,Ingresosv,Gastosfinv,Unv,fprecio,Utilidadesretenidasv,Impuestosv,Accionescv],axis=1)
altman.columns = ['Activo C','Activo', 'Pasivo C','Pasivo','Ingresos','Gastos Financieros','Utilidad Neta','Precio close','Utilidades R','Impuestos','#Acciones']

altman['Capitalizacion']= altman['Precio close']*altman['#Acciones']
altman['X1']=(altman['Activo C']-altman['Pasivo C'])/altman['Activo']
altman['X2']= altman['Utilidades R']/altman['Activo']
altman['X3']=(altman['Utilidad Neta']+ altman['Impuestos']+altman['Gastos Financieros'])/altman['Activo']
altman['X4']=altman['Capitalizacion']/altman['Pasivo']
altman['X5']=altman['Ingresos']/altman['Pasivo']
altman['Altman']=1.2*(altman['X1'])+1.4*(altman['X2'])+3.3*(altman['X3'])+0.6*(altman['X4'])+1*(altman['X5'])
altman['Indicador'] = ''

for i in range(len(altman)):
  if altman['Altman'][i] < 1.81:
    altman['Mensaje'][i] = 'Bien'
  if altman['Altman'][i] < 2.70:
    altman['Mensaje'][i] = 'Problemas'
  if altman['Altman'][i] < 2.99:
    altman['Mensaje'][i] = 'Seguridad'
  if altman['Altman'][i] < 2.99:
    altman['Mensaje'][i] = 'SOLVENTE'
altman=altman.transpose()
```

### **------------ WACC -----------**

```python
simbolo1 = '^GSPC'
ticker_object1 = yf.Ticker(simbolo1)

fechai1=fechas[0]-pd.DateOffset(years=5)
fechai2=fechas[1]-pd.DateOffset(years=5)
fechai3=fechas[2]-pd.DateOffset(years=5)

psyp1= ticker_object1.history(start=fechai1,end=fechas[0],interval='1mo')
psyp2= ticker_object1.history(start=fechai2,end=fechas[1],interval='1mo')
psyp3= ticker_object1.history(start=fechai3,end=fechas[2],interval='1mo')

psyp1 = psyp1["Close"]
psyp2 = psyp2["Close"]
psyp3 = psyp3["Close"]



psypt = pd.DataFrame({'Close1':psyp1,'Close2':psyp2,'Close3':psyp3})
psypt['Rendimiento1'] = (psypt['Close1']/psypt['Close1'].shift(1)-1)
psypt['Rendimiento2'] = (psypt['Close2']/psypt['Close2'].shift(1)-1)
psypt['Rendimiento3'] = (psypt['Close3']/psypt['Close3'].shift(1)-1)


rm1 = (psypt['Rendimiento1'].mean())*12
rm2 = (psypt['Rendimiento2'].mean())*12
rm3 = (psypt['Rendimiento3'].mean())*12

rmtot = {
    fechas[0]: rm1,
    fechas[1]: rm2,
    fechas[2]: rm3,
}

rmtot = pd.DataFrame.from_dict(rmtot, orient='index', columns=['rendimientos mesuales ultimos 5y S&P'])

print(rmtot)
```

```python
simbolo1 = '^GSPC'
ticker_object1 = yf.Ticker(simbolo1)

simbolo2 = 'BABA'
ticker_object2 = yf.Ticker(simbolo2)

#datos de syp ultimos 5 años cada año
fecham1=fechas[0]-pd.DateOffset(years=5)
fecham2=fechas[1]-pd.DateOffset(years=5)
fecham3=fechas[2]-pd.DateOffset(years=5)
psypd1= ticker_object1.history(start=fecham1,end=fechas[0],interval='1d')
psypd2= ticker_object1.history(start=fecham2,end=fechas[1],interval='1d')
psypd3= ticker_object1.history(start=fecham3,end=fechas[2],interval='1d')
psypd1 = psypd1["Close"]
psypd2 = psypd2["Close"]
psypd3 = psypd3["Close"]
psyptd = pd.DataFrame({'Close1':psypd1,'Close2':psypd2,'Close3':psypd3})
psyptd['R SYP 1'] = (psyptd['Close1']/psyptd['Close1'].shift(1)-1)
psyptd['R SYP 2'] = (psyptd['Close2']/psyptd['Close2'].shift(1)-1)
psyptd['R SYP 3'] = (psyptd['Close3']/psyptd['Close3'].shift(1)-1)


#datos de Alibaba ultimos 5 años cada año
pbaba1= ticker_object2.history(start=fecham1,end=fechas[0],interval='1d')
pbaba2= ticker_object2.history(start=fecham2,end=fechas[1],interval='1d')
pbaba3= ticker_object2.history(start=fecham3,end=fechas[2],interval='1d')

pbab1 = pbaba1["Close"]
pbab2 = pbaba2["Close"]
pbab3 = pbaba3["Close"]

pbabatot = pd.DataFrame({'Close1':pbab1,'Close2':pbab2,'Close3':pbab3})
pbabatot['R BABA 1'] = (pbabatot['Close1']/pbabatot['Close1'].shift(1)-1)
pbabatot['R BABA 2'] = (pbabatot['Close2']/pbabatot['Close2'].shift(1)-1)
pbabatot['R BABA 3'] = (pbabatot['Close3']/pbabatot['Close3'].shift(1)-1)


beta1=pd.concat([psyptd['R SYP 1'],pbabatot['R BABA 1']],axis=1)
beta1.dropna(subset=['R SYP 1','R BABA 1' ], inplace=True)
beta2=pd.concat([psyptd['R SYP 2'],pbabatot['R BABA 2']],axis=1)
beta2.dropna(subset=['R SYP 2','R BABA 2' ], inplace=True)
beta3=pd.concat([psyptd['R SYP 3'],pbabatot['R BABA 3']],axis=1)
beta3.dropna(subset=['R SYP 3','R BABA 3' ], inplace=True)


#sacar pendiente (beta)
from scipy.stats import linregress
slope1, intercept, r_value, p_value, std_err = linregress(beta1['R SYP 1'],beta1['R BABA 1'] )
slope2, intercept, r_value, p_value, std_err = linregress(beta2['R SYP 2'],beta2['R BABA 2'] )
slope3, intercept, r_value, p_value, std_err = linregress(beta3['R SYP 3'],beta3['R BABA 3'] )

beta1a=slope1
beta2a=slope2
beta3a=slope3

betas = {
    fechas[0]: beta1a,
    fechas[1]: beta2a,
    fechas[2]: beta3a,
}
betas = pd.DataFrame.from_dict(betas, orient='index', columns=['betas ultimos 5y'])

print(betas)
```

```python
wacc=pd.concat([Gastosfinv,Patrimoniov,Deudav,Uaiv,Impuestosv,rmtot,betas,rfs],axis=1)
wacc.columns = ['Intereses','Patrimonio', 'Deuda','UAI','Impuesto','Riesgo Mercado','Betas','Bonos EEUU']
wacc['%tax']=wacc['Impuesto']/wacc['UAI']
wacc['pesoDEUDA']=wacc['Deuda']/(wacc['Deuda']+wacc['Patrimonio'])
wacc['pesoPATRIMONIO']=wacc['Patrimonio']/(wacc['Deuda']+wacc['Patrimonio'])
wacc['Kd']=wacc['Intereses']/wacc['Deuda']
wacc['Ke']=wacc['Bonos EEUU']+ wacc['Betas']*(wacc['Riesgo Mercado']-wacc['Bonos EEUU'])
wacc['WACC']=(wacc['Ke']*wacc['pesoPATRIMONIO'])+(wacc['Kd']*(1-wacc['%tax'])*wacc['pesoDEUDA'])
wacc['WACC%']=wacc['WACC']*100
wacc=wacc.transpose()
```
